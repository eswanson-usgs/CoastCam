```python
#date picker
text= pn.widgets.TextInput(name="Selected Date", disabled=True)
date_picker = pn.widgets.DatePicker(name='Date Picker')

def display_date(dt):
    text.value = str(dt)
    return text

#version 1 - less complicated. One interaction without composing layout
interact(display_date, dt=date_picker)

#version 2 - more complicated but can compose elements in layout easier
# layout = pn.Column(date_picker, text)
# def update(event):
#     layout[1].object = display_date(date_picker.value)
# date_picker.param.watch(update, 'value')
# layout.servable()
```
